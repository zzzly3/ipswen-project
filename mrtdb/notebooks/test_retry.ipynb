```python
from pktest import *
import json
import numpy as np
from tqdm import tqdm
```

```python
ip = {}
with open('asip.jsonl') as f:
    for line in f:
        a = json.loads(line)
        ip[a[2]] = max(ip.get(a[2], -9999), a[1])
ip = list(ip.items())
np.random.shuffle(ip)
ip = ip[:100]
```

```python
runner = TestRunner()
runner.add_tests(TestGroupTTL.make_groups([i[0] for i in ip], 'tcp', 'all', ttl_min=5, ttl_max=30, max_retry=10))
with tqdm(total=runner.count_remain(), unit='tests') as pbar:
    while not runner.done():
        pbar.update(runner.run_batch())
```

```python
stats = {}
for k, p in runner.get_results():
    if p and p[-1] == k[0]:
        stats.setdefault(k[2], set()).add(k[0])
for i in stats:
    print(i, len(stats[i]), len(stats[i].intersection(stats['x1'])))
```
